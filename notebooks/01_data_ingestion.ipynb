Ingesting raw CSV datasets from AMFI and mfapi.in into raw data storage.

```python
import os
BASE_DIR = os.path.abspath(os.path.join(os.getcwd(), '..'))
if not os.path.exists(os.path.join(BASE_DIR, 'data', 'db')):
    BASE_DIR = os.getcwd()
os.chdir(BASE_DIR)
print("Working directory set to:", os.getcwd())

import os
import pandas as pd

BASE_DIR = os.path.abspath(os.path.join(os.getcwd(), '..'))
if not os.path.exists(os.path.join(BASE_DIR, 'data', 'raw')):
    BASE_DIR = os.getcwd()

print("Checking raw data files in:", os.path.join(BASE_DIR, 'data', 'raw'))
for fname in sorted(os.listdir(os.path.join(BASE_DIR, 'data', 'raw'))):
    if fname.endswith('.csv'):
        df = pd.read_csv(os.path.join(BASE_DIR, 'data', 'raw', fname))
        print(f"Loaded {fname} - Shape: {df.shape}")
```
