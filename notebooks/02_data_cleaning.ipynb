# Day 2: Data Cleaning and SQLite Schema Loading
Cleaning raw datasets, handling missing values via forward-filling, normalizing schemas, and loading into `bluestock_mf.db`.

```python
import os
BASE_DIR = os.path.abspath(os.path.join(os.getcwd(), '..'))
if not os.path.exists(os.path.join(BASE_DIR, 'data', 'db')):
    BASE_DIR = os.getcwd()
os.chdir(BASE_DIR)
print("Working directory set to:", os.getcwd())

import os
import sqlite3
import pandas as pd

BASE_DIR = os.path.abspath(os.path.join(os.getcwd(), '..'))
if not os.path.exists(os.path.join(BASE_DIR, 'data', 'db')):
    BASE_DIR = os.getcwd()

db_path = os.path.join(BASE_DIR, 'data', 'db', 'bluestock_mf.db')
print("Connecting to SQLite database at:", db_path)
conn = sqlite3.connect(db_path)

tables = pd.read_sql("SELECT name FROM sqlite_master WHERE type='table';", conn)
print("\nTables initialized in database:")
print(tables)

for table in tables['name']:
    count = pd.read_sql(f"SELECT COUNT(*) as count FROM {table};", conn)['count'].iloc[0]
    print(f"Table '{table}': {count} rows loaded.")
```
